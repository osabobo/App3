```python
# importing required packages for data manipulation

import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np
import seaborn as sns
```

```python
# extending the number of viewable columns
pd.options.display.max_columns = 30
```

```python
# loading the dataset
df = pd.read_csv("C:/Users/OSAGIE/Desktop/energydata_complete.csv")
df.head()
```

```python
# Applying MinMaxScaler to scale the features column 

from sklearn.preprocessing import MinMaxScaler

# instantiate the scaler
scaler = MinMaxScaler()
df=df[['T1','T6']]

# normalizing the features
nor_features = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# splitting data into features(independent) and target(dependent) variables

# predictors
features = nor_features.drop(['T6'], axis=1)

# target variable
target = nor_features['T6']
```

```python
# splitting data into training and testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import  LinearRegression
regr = LinearRegression()
regr.fit(X_train,y_train)
predicts=regr.predict(X_test)
print(y_test,predicts)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
round(regr.score(X_test,y_test),2)
```

The R^2 value 0.43

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicts)
round(mae, 2)#prints 0.063
```

```python
pred_lasso=lasso_reg.predict(X_test)
```

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,pred_lasso))
round(rmse, 3)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```
